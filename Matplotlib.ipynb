```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
```

```python
y=np.sin(x)
```

```python
z=np.cos(x)
```

```python
print (x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data
1.sine wave

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
# Adding the x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine Value')
plt.title('Sine Wave')
plt.show()
```

```python
# For cos graph
plt.plot(x,z)
plt.xlabel('Angle')
plt.ylabel('Cosine Value')
plt.title('Cos Wave')
plt.show()
```

```python
# Parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
# For Parabola
plt.plot(x,y)
plt.xlabel('Value of x')
plt.ylabel('Value of y corresponding to x')
plt.title('Parabola')
plt.show()
```
